```python
#define the functions needed, add, sub, mul, div
#print option to the user
#ask for values
#call the functions
#while loop to continue the program until the users want to exit

def add(a, b):
    answer = a + b
    print ( str(a) + " + " + str(b) + " = " + str(answer) + "\n")
    
def sub(a, b):
    answer = a - b
    print ( str(a) + " - " + str(b) + " = " + str(answer) + "\n")

def mul(a, b):
    answer = a * b
    print ( str(a) + " x " + str(b) + " = " + str(answer) + "\n")

def div(a, b):
    answer = a / b
    print ( str(a) + " : " + str(b) + " = " + str(answer) + "\n")
    
while True:
    print("A. Addition ")
    print("B. Subtraction ")
    print("C. Multiplication ")
    print("D. Division ")
    print("E. Exit")

    choice =  input("masukkan pilihan kamu ? ")

    if choice == "a" or choice == "A":
        print("Addition")
        a = int(input("input angka pertama :"))
        b = int(input("input angka kedua: "))
        add(a, b)
    elif choice == "b" or choice == "B":
        print("Subtraction")
        a = int(input("input angka pertama :"))
        b = int(input("input angka kedua: "))
        sub(a, b)
    elif choice == "c" or choice == "C":
        print("Multiplication")
        a = int(input("input angka pertama :"))
        b = int(input("input angka kedua: "))
        mul(a, b)
    elif choice == "d" or choice == "D":
        print("Division")
        a = int(input("input angka pertama :"))
        b = int(input("input angka kedua: "))
        div(a, b)
    elif choice == "e" or choice == "E":
        print("Exit")
        exit()
```

```python
#collect email from user
#split the email using the @, the second part is gonna be saved as domain
#split using .

def main():
    print("welcome to the email slicer")
    print("")
    
    email_input = input("masukkan email: ")
    (username, domain) = email_input.split("@")
    (domain, extension) = domain.split(".")
    
    print("Username: ",username)
    print("Domain: ", domain)
    print("Extension:", extension)

while True:
    main()
    
```
